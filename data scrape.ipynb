```python
def find_background(a):
    # soupp = BeautifulSoup(browser.page_source)
    # a = soupp.find('main',attrs = {'class':'core-rail'})
    #########################################################
    #   
    b = a.find('div',attrs = {'class':'ph5 pb5'}).find('ul',attrs = {'class':"pv-top-card--experience-list"})
    hasd = b.find_all('li')
    if len(hasd) >0:
        
        c = hasd[0].find('span')
        company = c.text.strip()
        try :
            d= b.find_all('li')[1].find('span').text.strip()
        except IndexError:
            d = None
        return (company,d)
    
    return(None,None)


def find_followers(a):
#     soupp = BeautifulSoup(browser.page_source)
#     a = soupp.find('main',attrs = {'class':'core-rail'})
    cfollowers = None
    e = a.find('div', attrs = {'class' : 'profile-detail'})
    ee = e.find_all('div',attrs = {'class' : 'pv-deferred-area ember-view'},recursive = False)
    for i in ee:
        eee = i.find_all('div',attrs = {'class':'pv-deferred-area__content'},recursive = False)
        #print(len(eee))
        if len(eee)>0:
            te = eee[0].find_all('header')
            #print(len(te))
            if len(te) >= 1:
                tet = te[0].find('h2')
                #print(tet.text.strip())
                if tet.text.strip() == 'Activity':
                    try:
                        follower = eee[0].find('span',attrs = {'class': 'align-self-center t-14 t-black--light'})
                        cfollowers = clean_follower(follower.text.strip())
                  #print(cfollowers)
                    except :
                        cfollowers = None

                        
    return cfollowers
    
def get_about(a,browser):
#     soupp = BeautifulSoup(browser.page_source)
#     a = soupp.find('main',attrs = {'class':'core-rail'})
    about = None
    pe = None
    browser.execute_script("document.body.style.zoom= '200%'")
    
    time.sleep(1)
    try: 
        print('started trying ')
        #browser.find_element_by_id('line-clamp-show-more-button').click()
        tt =  browser.find_element_by_id('line-clamp-show-more-button')
        browser.execute_script("arguments[0].setAttribute('data-artedco-is-focused','true')",tt)
        browser.execute_script("arguments[0].click()",tt)
        #print('successfully clicked')
        soupp = BeautifulSoup(browser.page_source)
        a = soupp.find('main',attrs = {'class':'core-rail'})
        e = a.find('div', attrs = {'class' : 'profile-detail'})
        ee = e.find_all('div',attrs = {'class' : 'pv-oc ember-view'},recursive = False)
        for i in ee:
            eee = i.find_all('section',recursive = False)
            #print(len(eee))
            if len(eee)>0:
                te = eee[0].find_all('header')  ## eee is a section
                #print(len(te))
                if len(te) >= 1:
                    tet = te[0].find('h2')
                    #print(tet.text.strip())
                    if tet.text.strip() == 'About':
                        about_ = eee[0].find('p').find('span',attrs = {'class': 'lt-line-clamp__raw-line'})
                        

                        about = about_.text.strip()
    
    except : #NoSuchElementException:
        e = a.find('div', attrs = {'class' : 'profile-detail'})
        ee = e.find_all('div',attrs = {'class' : 'pv-oc ember-view'},recursive = False)
        for i in ee:
            eee = i.find_all('section',recursive = False)
            #print(len(eee))
            if len(eee)>0:
                te = eee[0].find_all('header')  ## eee is a section
                #print(len(te))
                if len(te) >= 1:
                    tet = te[0].find('h2')
                    #print(tet.text.strip())
                    if tet.text.strip() == 'About':
                        about_ = eee[0].find('p')
                        if about_ != None :
                            about_ = about_.find_all('span',attrs = {'class': 'lt-line-clamp__line'})
                            abo = []
                            for i in about_: abo.append(i.text)

                            about = ' '.join(abo)
                        else :
                            about = None
                          #print(cfollowers)
    browser.execute_script("document.body.style.zoom= '30%'")
    return about
    

    
# def find_experience(a):
# #     soupp = BeautifulSoup(browser.page_source)
# #     a = soupp.find('main',attrs = {'class':'core-rail'})
#     try :    
#     ## Number of years

#         g = a.find('section', attrs = {'id':'experience-section'}).find('ul')
#         h =g.find_all('li')
#         duration = h[0].find('span', attrs = {'class':"pv-entity__bullet-item-v2"}).string
#         no_of_months = clean_duration(duration)

#     except Exception as e:
#         #print('this occured when finding number of months')
#         #print("<p>Error: %s</p>" % str(e) )
#         no_of_months = None

#     ## Location
#     try:
#         location = h[0].find('h4', attrs = {'class':"pv-entity__location t-14 t-black--light t-normal block"}).find_all('span')[-1].string
#     except Exception as e:
#         location = None

#     return (no_of_months,location)

def find_experience(a):
#     soupp = BeautifulSoup(browser.page_source)
#     a = soupp.find('main',attrs = {'class':'core-rail'})
    try :    
    ## Number of years

        g = a.find('section', attrs = {'id':'experience-section'}).find('ul')
        h =g.find_all('li')
        duration = h[0].find('span', attrs = {'class':"pv-entity__bullet-item-v2"}).string
        no_of_months = clean_duration(duration)
       
        Company_link = 'https://www.linkedin.com'+h[0].find('a')['href']
        
        
    except Exception as e:
        #print('this occured when finding number of months')
        #print("<p>Error: %s</p>" % str(e) )
        no_of_months = None
        Company_link = None

    ## Location
    try:
        location = h[0].find('h4', attrs = {'class':"pv-entity__location t-14 t-black--light t-normal block"}).find_all('span')[-1].string
    except Exception as e:
        location = None

    return (no_of_months,location,Company_link)

def get_info(browser):
    soupp = BeautifulSoup(browser.page_source)
    a = soupp.find('main',attrs = {'class':'core-rail'})
    company,school = find_background(a)
    followers = find_followers(a)
    no_of_months,location,company_link = find_experience(a)
    about = get_about(a,browser)
    return [company,school,followers,no_of_months,location,company_link,about]

def segment(n,i,urls):
    return urls[n*(i-1):n*i]


def run_parralell(browsers,list_urls):
    ams = []
    for i in list(zip(browsers,list_urls)):
        browser,link = i
        browser.get(link)
        time.sleep(1)
        browser.execute_script("document.body.style.zoom= '30%'")
    time.sleep(15)  ## time for each browser to load page
   
    for i in list(zip(browsers,list_urls)) :
        browser,link = i
        dat = get_info(browser)
        dat.append(link)
        ams.append(dat)
    return ams


```
